```python
import os
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown,display
```

```python
load_dotenv(override=True)
openai_api_key = os.getenv('OPENAI_API_KEY')
anthropic_api_key = os.getenv('ANTHROPIC_API_KEY')
google_api_key = os.getenv('GOOGLE_API_KEY')

anthropic_url = "https://api.anthropic.com/v1/"
gemini_url = "https://generativelanguage.googleapis.com/v1beta/openai/"

anthropic = OpenAI(api_key=anthropic_api_key, base_url=anthropic_url)
gemini = OpenAI(api_key=google_api_key, base_url=gemini_url)



openai=OpenAI()
anthropic = OpenAI(api_key=anthropic_api_key, base_url=anthropic_url)
gemini = OpenAI(api_key=google_api_key, base_url=gemini_url)
```

```python
gpt_model="gpt-4.1-mini"
gemini_model="gemini-2.5-flash-lite"
claude_model="claude-3-5-haiku-20241022"
```

```python
gpt_system="you are an angry teacher who’s tired and frustrated with your students Alex and Charlie."
gemini_system="You are Alex, you and your friend Charlie are stupid students who don’t understand anything."
claude_system="You are Charlie, you and your friend Alex are stupid students who don’t understand anything."

gpt_messages=["Hi everybody"]
gemini_messages=["Hello dad!"]
claude_messages=["hello mom!"]
```

```python
conversation=""
conversation+=f"""teacher : {gpt_messages[0]}\n"""
conversation+=f""" Alex: {gemini_messages[0]}\n"""
conversation+=f"""Charlie:{claude_messages[0]}\n"""
```

```python
def call_gpt(conversation):
    user_prompt=f"""You are teacher, in conversation with Alex and Charlie.
    The conversation so far is as follows:
    {conversation}
    Now with this, respond with what you would like to say next, as teacher.
    answer like :
    Teacher :......
    """

    messages=[{"role":"system","content":gpt_system},
    {"role":"user","content":user_prompt}]
    response=openai.chat.completions.create(model=gpt_model,messages=messages)
    return response.choices[0].message.content
    
```

```python
call_gpt(conversation)
```

```python
def call_gemini(conversation):
    user_prompt=f"""You are Alex, in conversation with teacher and Charlie.
    The conversation so far is as follows:
    {conversation}
    Now with this, respond with what you would like to say next, as Alex.
    answer like :
    Alex :......
    """
    messages=[{"role":"system","content":gemini_system},
    {"role":"user","content":user_prompt}]
    response=gemini.chat.completions.create(model=gemini_model,messages=messages)
    return response.choices[0].message.content
```

```python
call_gemini(conversation)
```

```python
def call_claude(conversation):
    user_prompt=f"""You are Charlie, in conversation with teacher and Alex.
    The conversation so far is as follows:
    {conversation}
    Now with this, respond with what you would like to say next, as Charlie.
    answer like :
    Charlie :......
    """
    messages=[{"role":"system","content":claude_system},
    {"role":"user","content":user_prompt}]
    response=anthropic.chat.completions.create(model=claude_model,messages=messages)
    return response.choices[0].message.content
```

```python
 

def chat_models(conversation):
    display(Markdown(conversation))
    for i in range(5):
        next_gpt=call_gpt(conversation)
        display(Markdown(next_gpt))
        gpt_messages.append(next_gpt)
        conversation+=f"""teacher :{next_gpt}\n"""

        next_gemini=call_gemini(conversation)
        display(Markdown(next_gemini))
        gemini_messages.append(next_gemini)
        conversation+=f"""Alex :{next_gemini}\n"""

        next_claude=call_claude(conversation)
        display(Markdown(next_claude))
        claude_messages.append(next_claude)
        conversation+=f"""Charlie :{next_claude}\n"""

```

```python
chat_models(conversation)
```
